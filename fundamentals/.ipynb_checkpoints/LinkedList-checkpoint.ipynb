```python
# =======creation of linked list===========

class Node:
    def __init__(self, dataval=None):
        self.dataval = dataval
        self.nextval = None

class SLinkedList:
    def __init__(self):
        self.headval = None
    
list1 = SLinkedList()
list1.headval = Node("Mon")
e2 = Node("Tue")
e3 = Node("Wed")
# Link first Node to second node
list1.headval.nextval = e2

# Link second Node to third node
# e3 is a Node object, assign its address to e2 nextval
e2.nextval = e3
```

```python
# ========traverse to print all data in the linked list========

class Node:
    def __init__(self, dataval=None):
        self.dataval = dataval
        self.nextval = None

class SLinkedList:
    def __init__(self):
        self.headval = None

    def listprint(self):
        # printval is headnode, name represents its address
        printval = self.headval
        while printval is not None: # addr is valid
            print (printval.dataval)
            # use variable printval to traverse the linked list,
            # till getting all data in the linked list
            printval = printval.nextval
            
# create a singly linked list named list
list = SLinkedList()
list.headval = Node("Mon")
e2 = Node("Tue")
e3 = Node("Wed")

# Link first Node to second node
list.headval.nextval = e2

# Link second Node to third node
e2.nextval = e3

'''
finally, the elements in the linked list sorted like this:
headval -> e2 -> e3 -> None
and here, these variable names indicate memory location
'''
list.listprint()
```

```python
#=========Inserting at the Beginning of the Linked List==========

class Node:
    def __init__(self, dataval=None):
        self.dataval = dataval
        self.nextval = None

class SLinkedList:
    def __init__(self):
        self.headval = None

    # ========traverse to print all data in the linked list========
    def listprint(self):
        # printval is headnode, name represents its address
        printval = self.headval
        while printval is not None: # addr is valid
            print (printval.dataval)
            # use variable printval to traverse the linked list,
            # till getting all data in the linked list
            printval = printval.nextval

    # =========Insertion in a Linked List==========
    '''
    There are three cases to insert a new element in the linked list
    beginning, middle and end of the linked list
    '''


    def InsertAtBegining(self,newdata):
        NewNode = Node(newdata) # write newdata to the constructor dataval

    # Update the new nodes next val to existing node
        NewNode.nextval = self.headval  # since insert at the beginning, pointer nextval points to headval
        self.headval = NewNode  # now the headval change to be NewNode

list = SLinkedList()
list.headval = Node("Mon")
e2 = Node("Tue")
e3 = Node("Wed")

list.headval.nextval = e2
e2.nextval = e3

list.InsertAtBegining("Sun")

list.listprint()
```

```python
# Inserting at the End of the Linked List

class Node:
    def __init__(self, dataval=None):
        self.dataval = dataval
        self.nextval = None

class SLinkedList:
    def __init__(self):
        self.headval = None

# Function to add newnode

    def InsertAtEnd(self, newdata):
        NewNode = Node(newdata)
        
        # if headval is empty
        # just point headval to NewNode
        # the nextval of NewNode is None because of constructor
        if self.headval is None:
            self.headval = NewNode
            return
        
        laste = self.headval
        # traverse from headnode to find the last node
        # the end condition is that the nextval equals to None
        while(laste.nextval):
            laste = laste.nextval
        laste.nextval = NewNode

# Print the linked list
    def listprint(self):
        printval = self.headval
        while printval is not None:
            print (printval.dataval)
            printval = printval.nextval


list = SLinkedList()
list.headval = Node("Mon")
e2 = Node("Tue")
e3 = Node("Wed")

list.headval.nextval = e2
e2.nextval = e3

list.InsertAtEnd("Thu")

list.listprint()
```

```python
# Inserting in between two Data Nodes

class Node:
    def __init__(self, dataval=None):
        self.dataval = dataval
        self.nextval = None

class SLinkedList:
    def __init__(self):
        self.headval = None

# Function to add node
# insert after tobeinserted_node
    def InsertInbetween(self,tobeinserted_node,newdata):
        if tobeinserted_node is None:
            print("The mentioned node is absent")
            return
        
        # data of the new inserted node
        NewNode = Node(newdata)
        # next pointer of the new inserted node
        NewNode.nextval = tobeinserted_node.nextval
        tobeinserted_node.nextval = NewNode

# Print the linked list
    def listprint(self):
        printval = self.headval
        while printval is not None:
            print (printval.dataval)
            printval = printval.nextval


list = SLinkedList()
list.headval = Node("Mon")
e2 = Node("Tue")
e3 = Node("Thu")

list.headval.nextval = e2
e2.nextval = e3

list.InsertInbetween(list.headval.nextval,"Fri")

list.listprint()
```

```python
# Removing an Item form a Liked List

class Node:
    def __init__(self, data=None):
        self.data = data
        self.next = None

class SLinkedList:
    def __init__(self):
        self.head = None

    def InsertAtbegining(self, data_in):
        NewNode = Node(data_in)
        NewNode.next = self.head
        self.head = NewNode

# Function to remove node

    def RemoveNode(self, Removekey):

        HeadVal = self.head

        if (HeadVal is not None):
            if (HeadVal.data == Removekey):
                self.head = HeadVal.next
                HeadVal = None
                return
            
        # traverse till find the removekey
        while (HeadVal is not None):
            if HeadVal.data == Removekey:
                break
                
            # after the loop, prev points to the former one node 
            # that removekey (data) corresponding to
            prev = HeadVal
            HeadVal = HeadVal.next

        if (HeadVal == None):
            return

        prev.next = HeadVal.next

        HeadVal = None

    def LListprint(self):
        printval = self.head
        while (printval):
            print(printval.data)
            printval = printval.next


llist = SLinkedList()
llist.InsertAtbegining("Mon")
llist.InsertAtbegining("Tue")
llist.InsertAtbegining("Wed")
llist.InsertAtbegining("Thu")
llist.RemoveNode("Tue")
llist.LListprint()
```
